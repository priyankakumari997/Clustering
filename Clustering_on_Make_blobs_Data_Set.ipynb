```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=1000,centers=5,n_features=4,random_state=2526)
X.shape, y.shape
```

```python
import seaborn as sns
import pandas as pd
sns.pairplot(pd.DataFrame(X))
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5,init='k-means++',n_init=10,max_iter=300,random_state=250)
kmeans.fit(X)
kmeans.cluster_centers_
```
